```python
import pandas as pd
```

```python
data_1 = pd.read_csv('/Users/jbian/Downloads/download_data/1.csv',converters={i: str for i in range(0, 100)})
data_2 = pd.read_csv('/Users/jbian/Downloads/download_data/2.csv',converters={i: str for i in range(0, 100)})
data_3 = pd.read_csv('/Users/jbian/Downloads/download_data/3.csv',converters={i: str for i in range(0, 100)})
data_4 = pd.read_csv('/Users/jbian/Downloads/download_data/4.csv',converters={i: str for i in range(0, 100)})
data_5 = pd.read_csv('/Users/jbian/Downloads/download_data/5.csv')
data_6 = pd.read_csv('/Users/jbian/Downloads/download_data/6.csv',converters={i: str for i in range(0, 100)})
data_7 = pd.read_csv('/Users/jbian/Downloads/download_data/7.csv')
```

```python
print(data_1.head())
print(data_2.head())
print(data_3.head())
data_4.columns = ['id','msno','lang_repeat_top1','lang_repeat_top2','lang_repeat_top3']
print(data_4.head())
print(data_5.head())
print(data_6.head())
print(data_7.head())
```

```python
data_merge_1 = data_1.merge(data_2[['msno','genre_repeat_top1','genre_repeat_top2','genre_repeat_top3',
                    'genre_repeat_top4','genre_repeat_top5','genre_repeat_top6',
                    'genre_repeat_top7','genre_repeat_top8']], on = 'msno')
```

```python
data_merge_1.head()
```

```python
data_merge = data_merge_1.merge(data_3[['msno','lang1','lang2','lang3']], on = 'msno')\
                         .merge(data_4[['msno','lang_repeat_top1','lang_repeat_top2',
                                       'lang_repeat_top3']], on='msno')\
                         .merge(data_5[['msno','song_counts','genre_counts']], on = 'msno')\
                         .merge(data_6[['msno','artist_name1','artist_name2',
                                       'artist_name3']], on = 'msno')\
                         .merge(data_7[['msno','mean_song_len']],on='msno')
```
